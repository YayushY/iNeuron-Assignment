```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
```

```python
df
```

```python
df["FlightNumber"]=df["FlightNumber"].fillna(method="ffill",downcast='int64')
```

```python
df
```

```python
df["FlightNumber"] = [10045,10055,10065,10075,10085]
```

```python
df["From"] = [0,0,0,0,0]
df["To"] = [0,0,0,0,0]
```

```python
df
```

```python
for i in range(len(df["From_To"])):
    k=df["From_To"][i].split("_")
    df["From"][i]=k[0]
    df["To"][i] = k[1]
```

```python
df
```

```python
df.drop(["From_To"],axis=1)
```

```python
df=df[["From","To","FlightNumber","Airline","RecentDelays"]]
df
```

```python
df["From"]=df["From"].str.capitalize()
df["To"] = df["To"].str.capitalize()
df
```

```python
df["delay1"] = [0,0,0,0,0]
df["delay2"] = [0,0,0,0,0]
```

```python
df["delay3"] = [0,0,0,0,0]
    
```

```python
df
```

```python
for i in range(len(df["RecentDelays"])):
    if(len(df["RecentDelays"][i])==0):
        df["delay1"][i]=np.nan
        df["delay2"][i]=np.nan
        df["delay3"][i]=np.nan
    if(len(df["RecentDelays"][i])==1):
        df["delay1"][i]=df["RecentDelays"][i][0]
        df["delay2"][i]=np.nan
        df["delay3"][i]=np.nan
    if(len(df["RecentDelays"][i])==2):
        df["delay1"][i]=df["RecentDelays"][i][0]
        df["delay2"][i]=df["RecentDelays"][i][1]
        df["delay3"][i]=np.nan
    if(len(df["RecentDelays"][i])==3):
        df["delay1"][i]=df["RecentDelays"][i][0]
        df["delay2"][i]=df["RecentDelays"][i][1]
        df["delay3"][i]=df["RecentDelays"][i][2]
```
